```python
import numpy as np
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data=np.loadtxt("boston_train", delimiter=",")
data=scaler.fit_transform(data)
```

```python
def step_gradient(pt, lr, m, c):
    m_slope=0
    c_slope=0
    for i in range(M):
        x=pt[i,0]
        y=pt[i,1]
        m_slope+=(-2/M)*(y-m*x-c)*x
        c_slope+=(-2/M)*(y-m*x-c)
        new_m=m-lr*m_slope
        new_c=c-lr*c_slope
    return new_m, new_c
```

```python
def gradient(pt,lr,num_iter):
    m=0
    c=0
    for i in range(num_iter):
        m,c=step_gradient(pt,lr,m,c)
    return m,c
```

```python
def cost(pts,m,c):
    total_cost=0
    M=len(pts)
    for i in range(M):
        x=pts[i,0]
        y=pts[i,1]
        total_cost+=(1/M)*((y-m*x-c)**2)
    return total_cost
```
